```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
weather1 = pd.read_csv("C:/Users/HP/Downloads/seattle_weather.csv")
weather2 = pd.read_csv("C:/Users/HP/Downloads/austin_weather.csv")
weather1.head(3)
```

```python
austin_weather = weather2
month_mapping = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
austin_weather['MONTH'] = austin_weather['DATE'].map(month_mapping)
austin_weather.set_index("DATE")
austin_weather.head(2)
```

```python
seattle_weather = weather1[weather1['STATION']=='USW00094290'].copy()

# Mapping dictionary for month names
month_mapping = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
seattle_weather['MONTH'] = seattle_weather['DATE'].map(month_mapping)
seattle_weather = seattle_weather.set_index('DATE')
seattle_weather.head(3)
```

#### PY-PLOT Interface

```python
fig, ax = plt.subplots()
```

## a. Adding data to axes

```python
fig , ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'])
plt.show()
```

```python
fig , ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'])
ax.plot(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'])
plt.show()
```

## b. Customizing plots

### Adding Markers
Add markers to show data is NOT continuous.
Adding markers would show us where the data exists and which parts are just lines that connect between the data points.

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], marker = 'o')
plt.show()
```

### Setting the linestyle
You can change the appearance of the connecting lines to further emphasize that the data is NOT continuous.

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], marker = 'o', linestyle = '--')
plt.show()
```

### Eliminating lines with linestyle

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], marker = 'v', linestyle = 'None')
plt.show()
```

### Set colour for the data

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], marker = 'o', linestyle = '--', color = 'g')
plt.show()
```

### Customize the axis labels and add Title

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker = 'o', color = 'c')
ax.set_xlabel('Time (Months)')
ax.set_ylabel('Average Temperatue (Fahrenheit Degrees)')
ax.set_title('Weather in Seattle')
plt.show()
```

## c. Small Multiples

```python
# figure object with 2 rows of sub plots, and 1 columns.
# This is a one dimensional array, use one index to access the elements.
# 2 rows = [0,1]
fig, ax = plt.subplots(2,1) 
plt.show()
```

### Adding error bars to a line plot

```python
fig, ax = plt.subplots()
ax.errorbar(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], yerr = seattle_weather['MLY-TAVG-STDDEV'])
ax.errorbar(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'], yerr = austin_weather['MLY-TAVG-STDDEV'])
ax.set_ylabel('Temperature (Fahrenheit)')
```
