```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import median
```

# 3. Visualizing a Categorical and a Quantitative Variable

Use catplot() which has the same advatages as a relplot()
- Shows the distribution of a quantitative variable within categories defined by a categorical variable.
- Example: Bar plots, Count plots, Box plots, point plots

### a. Count plots and bar plots (Categorical Plots)

```python
tips = sns.load_dataset('tips')
tips.head()
```

- Bar plots show the mean of the variable in each category.
- Lines in the following bar plot shows the 95% confidence interval.
- It means we are 95% sure that the true population mean lies within the confidence interval shown.
- To remove C.I set ci = None

```python
sns.catplot(data = tips, x = 'day', y = 'tip', kind = 'bar')
```

### b. Box Plot

```python
#showfliers = False removes the outliers
a = sns.catplot(x = 'time', y = 'total_bill', data = tips, kind = 'box', order = ['Dinner','Lunch'], showfliers = False)
```

### c. Point Plot

```python
sns.catplot(data=tips, x = 'smoker', y = 'total_bill', kind = 'point', estimator = median, linestyle='none', capsize=0.2)
# to remove confidence interval, set C.I = None
```

# 4. Customizing sea-born plots

```python
# darkgrid, whitegrid, dark, white, ticks}
sns.set_style('darkgrid')
sns.catplot(x = 'time', y = 'total_bill', data = tips, kind = 'box', order = ['Dinner','Lunch'], showfliers = False)
```

#### Set palette colors
