```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# 1. Introduction to seaborn

```python
# Create a scatter plot
height = [62,64,67,76,65,69,70,73]
weight = [120,136,148,175,137,165,154,172]

sns.scatterplot(x=height, y=weight)
```

```python
# Create a count plot
gender = ['Female','Female','Female','Female','Male','Male','Male','Male','Male','Male',]
sns.countplot(x = gender)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Using Hue

```python
hue_colors = {'Yes':'b','No':'c'}
sns.scatterplot(data=tips, x = 'total_bill', y='tip', hue='smoker', hue_order=['Yes','No'],palette=hue_colors)
plt.show()
```

```python
sns.countplot(data = tips, x= 'smoker', hue = 'sex')
plt.show()
```

# 2. Visualizing two quantitative variables

## a. Rel plots
- relplots() helps to create subplots in a single figure
- Show the relationship between two quantitative variables.
- Example: Scatter plots, line plots

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'smoker') 
#use col to create subplots horizontally
#use row to create subplots vertically
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'smoker', row = 'time')
```

```python
# use col_wrap=2 to display only 2 column on a single line
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'day',col_order=['Sat','Sun','Thur','Fri'])
```

## b. Customizing scatter plots

### 1. Point size and hue

```python
tips.head()
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'size', size = 'size')
```

### 2. Sub groups with point style

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'smoker', style = 'smoker')
```

### 3. Changing point transparency

```python
# Set alpha to be between 0 and 1
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter', alpha =0.5)
```
