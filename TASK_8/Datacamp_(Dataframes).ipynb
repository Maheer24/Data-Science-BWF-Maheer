```python
import pandas as pd
import numpy as np
```

# Transforming Dataframes

```python
dogs = {'Name':['Bella','Charlie','Lucy','Cooper','Max','Stella','Bernie'],
        'Breed': ['Labrador','Poodle','Chow Chow','Schnauzer','Labrador','Chihuahua','St. Bernard'],
        'Color': ['Brown','Black','Brown','Gray','Black','Tan','White'],
        'Height (cm)': [56,43,46,49,59,18,77],
        'Weight (kg)': [25,23,22,17,29,2,74],
        'Date of Birth': ['2013-07-01','2016-09-16','2014-08-25','2011-12-11','2017-01-20','2015-04-20','2018-02-27']
       }

dogs_df = pd.DataFrame(dogs)
dogs_df
```

```python
columns_to_rename = {col: col.replace(' ', '_').lower()  for col in dogs_df.columns}
dog = dogs_df.rename(columns = columns_to_rename)
dog
```

## a. Sorting

```python
# sort by column -> acsending
dog.sort_values('height_(cm)')
```

```python
# sort by column -> descending order
dog.sort_values('weight_(kg)',ascending=False)
```

```python
# sorting by multiple columns
dog.sort_values(['height_(cm)','weight_(kg)'],ascending = [True,False])
```

## b. Subsetting

```python
# find all the dogs whose height is greater than 55 cm
dog[dog['height_(cm)'] > 55]
```

```python
# subsetting based on text data
dog[dog['breed'] == 'Labrador']

# use .isin() 
is_labrador = dog['breed'].isin(['Labrador'])
dog[is_labrador]
```

```python
# subsetting based on dates
dog[dog["date_of_birth"] < "2015-01-01"]
```

```python
# subsetting using .isin() 
is_gray_or_black = dog['color'].isin(['Gray','Black'])
dog[is_gray_or_black]
```

## c. Adding a new column

```python
dog['height_(m)'] = dog['height_(cm)'] / 100
dog
```

```python
dog['bmi'] = dog['weight_(kg)'] / (dog['height_(m)'] ** 2)
dog.head()
```

## d. Multiple Manipulations

```python
# filter dogs with bmi less than 100
bmi_100 = dog[dog['bmi'] < 100]
# filter w.r.t heights in descending order
bmi_height_100 = bmi_100.sort_values('height_(cm)',ascending = False)

bmi_height_100[['name','height_(cm)','bmi']]
```

# Aggregating dataframes 

## a. Summary statistics

```python
# some summary statistics

#.mean(), .median(), .mode()
#.min(), .max()
#.var(), .std()
#.sum(), .quantile()
```

```python
dog['weight_(kg)'].sum()
```

```python
# oldest dog DOB
dog['date_of_birth'].min()
```

```python
# .agg() method is used to compute custom statistics

# the following function calculates the 30th percentile of a dataframe column
def pct30(column):
    return column.quantile(0.3)

def pct40(column):
    return column.quantile(0.4)

dog['weight_(kg)'].agg(pct30)
```

```python
# .agg() method can also be called on multiple columns
dog[['weight_(kg)','height_(cm)']].agg(pct30)
```

```python
# .agg() method can also be used to calculate multiple summary statistics
dog[['weight_(kg)','height_(cm)']].agg([pct30,pct40,'median'])
```

## b. Cumulative Statistics

```python
# Pandas also has methods for cumulative statistics
# Cumulative statistic functions, return a whole column.
# .cumsum(), .cummax(), .cummin(), .cumprod()

dog['weight_(kg)'].cumsum()
```

## c. Counting

```python
vet_visit = {'date': ['2013-07-01','2013-09-16','2013-08-25','2013-12-11','2013-07-01','2013-01-20','2013-04-20','2013-02-27','2013-07-09','2013-09-29','2013-01-05','2013-07-18'],
      'name':['Bella','Charlie','Lucy','Bella','Cooper','Max','Stella','Bernie','Bella','Charlie','Max','Cooper'],
      'breed': ['Labrador','Poodle','Chow Chow','Labrador','Schnauzer','Labrador','Chihuahua','St. Bernard','Beagle','Poodle','Poodle','Schnauzer'],
      'weight_kg': [25.9,23.1,22,25.9,17.3,29,2.8,74,25.9,23.1,29,17.3],
      'height_cm': [75,50,22,40,56,60,30,74,75,50,60,56],
      'color': ['Brown','Black','Brown','Brown','Gray','Black','Tan','White','Brown','Black','Black','Gray'],
       }
df = pd.DataFrame(vet_visit)
df
```

### Dropping duplicate pairs

```python
df.drop_duplicates(subset = ["name","breed"])
```

### Count no. of dogs of each breed

```python
# setting sort = True will put the breeds with biggest count on top
df['breed'].value_counts(sort = True)
```

```python
# normalize argument will turn the counts into proportion of the total
df['breed'].value_counts(normalize = True)
```

## d. Grouped summary statistics

### Grouped summaries

```python
# mean weight for each dog breed
df.groupby('breed')['weight_kg'].mean()
```

### Multiple grouped summaries

```python
df.groupby('breed')['weight_kg'].agg(['min','max','sum'])
```

### Grouping by multiple variables

```python
# Following code gives us the mean weight of each breed of each color
df.groupby(['color','breed'])['weight_kg'].mean()
```

```python
# har color aur har breed kay dog ka mean weight aur height 
df.groupby(['color','breed'])[['weight_kg','height_cm']].mean()
```

## e. Pivot tables

```python
df.pivot_table(index = 'color', values = 'weight_kg', aggfunc = ['mean','median'])
```

```python
# To group by two variables pass the second column in the columns argument (first argument is index)

# we did not pass argfunc in below code because pivot table calculates mean by default

# fill missing values by using fill_value argument

# if we set margins = True the last row and last column will contain the mean of the column or row(excluding missing values i.e. 0)
df.pivot_table(index = 'color',columns = 'breed', values = 'weight_kg', fill_value = 0, margins = True)
```

# Slicing and Indexing dataframes

## a. Explicit Indexes

```python
# Setting a column as the index
dogs_ind = dog.set_index('name')
dogs_ind.head()
```

```python
# Removing an index
dogs_ind.reset_index()
```

```python
# index values don't need to be unique
dogs_ind2 = dog.set_index('breed')
dogs_ind2.head()
```

```python
# multi level indexes
dog_ind3 = dog.set_index(['breed','color'])
dog_ind3.head()
```

```python
# subset inner levels with a list of tuples
dog_ind3.loc[[("Labrador","Brown"),("Chihuahua","Tan")]]
```
